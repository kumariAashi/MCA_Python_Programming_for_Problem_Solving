```python
# Program to search for an item in a list

# Input
items = [12, 45, 33, 21, 88]
search_item = 33

# Search operation
if search_item in items:
    position = items.index(search_item) + 1  # +1 for human-friendly position
    print(f"Item found at position {position}")
else:
    print("Item not found.")
```
