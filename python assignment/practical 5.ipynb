```python
#Given a list of employee records as dictionaries, sort them by salary and
# display the sorted list.

#Take input from user
employees = eval(input("Enter list of employee records: "))
print("Enter list of employee records: ",employees)
```

```python
#Sort by salary using lambda

sorted_list = sorted(employees, key=lambda x: x["salary"])
print("Sorted by Salary:", sorted_list)
```
