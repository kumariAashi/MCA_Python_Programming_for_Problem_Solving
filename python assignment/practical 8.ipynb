```python
#Write a Python program that lists all files in a directory and categorizes them by file extension.

# Input list of files
files = ['report.pdf', 'data.csv', 'photo.jpg', 'summary.pdf', 'script.py']

# Dictionary to store files by extension
file_types = {}

# Categorize files
for file in files:
    # Split filename and extension
    name, extension = file.split(".")
    
    # Add file under its extension
    if extension not in file_types:
        file_types[extension] = [file]
    else:
        file_types[extension].append(file)

# Display categorized files
print("Files categorized by extension:")
for ext, names in file_types.items():
    print(f".{ext} : {names}")
```
