```python
#Number triangle

for i in range(1, 6):
    for j in range(1, i + 1):
        print(j, end=" ")
    print()
```

```python
#Triangle (Star Pyramid)

rows = int(input("Enter number of rows: "))

for i in range(1, rows + 1):
    print(" " * (rows - i) + "* " * i)
```

```python
#Diamond shape

# Upper part
for i in range(1, 4):
    print(" " * (3 - i) + "* " * i)
# Lower part
for i in range(2, 0, -1):
    print(" " * (3 - i) + "* " * i)
```

```python
#Hollow Square Star Pattern

n = int(input("Enter size: "))
print("Enter size: ",n)

for i in range(n):
    for j in range(n):
        if i == 0 or i == n-1 or j == 0 or j == n-1:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
```

```python
#Cross “X” Shape

n = int(input("Enter size: "))
print("Enter size: ",n)

for i in range(n):
    for j in range(n):
        if i == j or i + j == n - 1:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
```
