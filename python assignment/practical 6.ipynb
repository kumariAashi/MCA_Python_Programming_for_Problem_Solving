```python
#Write a program that reads a text file and counts the number of lines, words, and characters.

# Create a sample file
with open("sample.txt", "w") as f:
    f.write("Hello World!\nPython makes programming fun.")
print("sample.txt file created!")
```

```python
#Read file and count lines, words, and characters

# Initialize counters
lines = words = chars = 0

# Read and process the file
with open("sample.txt", "r") as f:
    for line in f:
        lines += 1
        words += len(line.split())
        chars += len(line)

# Display results
print("Lines:", lines)
print("Words:", words)
print("Characters:", chars)

```
