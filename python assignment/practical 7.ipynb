```python
#Read a sentence and display how many times each word appears, ignoring case and punctuation

#Take input from user

text = input("Enter a sentence: ")
print("Enter a sentence:",text)
```

```python
#Clean the text (remove punctuation and convert to lowercase)
for ch in ".,!?":
    text = text.replace(ch, "")
text = text.lower()

#Count word frequency
words = text.split()
freq = {}

for word in words:
    freq[word] = freq.get(word, 0) + 1
    
#Display result
print(freq)
```
