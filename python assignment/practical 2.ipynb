```python
#wap to find the probabilty of rolling a dice


# total possible outcomes on a dice
total_outcomes = 6

# take user input for desired outcome
desired = int(input("Enter the number you want (1 to 6): "))
print("Enter the number you want (1 to 6): ",desired)
```

```python
# check if input is valid
if 1 <= desired <= 6:
    probability = 1 / total_outcomes
    print(f"Probability of getting {desired} is: {probability}")
else:
    print("Invalid number! Please enter a number between 1 and 6.")
```

```python
#wap to find the probabilty of flipping a point

# total possible outcomes when flipping a coin
total_outcomes = 2   # (Head or Tail)

# favorable outcomes for getting Head
favorable_outcomes = 1

# probability formula
probability = favorable_outcomes / total_outcomes

# display result
print("Probability of getting a Head:", probability)
print("Probability of getting a Tail:", probability)
```
